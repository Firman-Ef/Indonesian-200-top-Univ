```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IndonesiaTopUniv2022.csv")
```

### 2022 Indonesian University Ranking

```python
df.head()
```

```python
df.rename(columns={'#Rank':'Rank'}, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
# Top 10 Indonesian University by Rank
top_University = df.sort_values(by='Rank', ascending=True)[:10]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=top_University.University, x=top_University.Rank)
plt.xticks()
plt.xlabel('Rank')
plt.ylabel('University')
plt.title('2022 Top 10 Indonesian University by Rank')
plt.show()
figure.savefig('uniranktop10.png', bbox_inches='tight')
```

```python
#2nd Top 10 Indonesian University by Rank
top_University = df.sort_values(by='Rank', ascending=True)[10:20]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=top_University.University, x=top_University.Rank)
plt.xticks()
plt.xlabel('Rank')
plt.ylabel('University')
plt.title('2nd Top 10 Indonesian University by Rank')
plt.xlim([10, 21])
plt.show()
figure.savefig('uniranktop2nd10.png', bbox_inches='tight')
```

```python
# Least 10 Indonesian University by Rank
top_University = df.sort_values(by='Rank', ascending=True)[568:578]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=top_University.University, x=top_University.Rank)
plt.xticks()
plt.xlabel('Rank')
plt.ylabel('University')
plt.title('Least 10 Indonesian University by Rank')
plt.xlim([568, 578])
plt.show()
figure.savefig('unirankleast10.png', bbox_inches='tight')
```
